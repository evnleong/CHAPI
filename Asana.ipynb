```python
#import all libraries
import requests
from requests.auth import HTTPBasicAuth
import json
import pandas as pd
import csv
import os
from dotenv import load_dotenv
```

```python
starrez_url = "https://cornell.starrezhousing.com/StarRezRest/services/"
```

```python
load_dotenv(".env")
basic = HTTPBasicAuth(os.environ.get('STARREZ_USERNAME'), os.environ.get('STARREZ_TOKEN'))
headers =  {"accept": "application/json","content-type": "application/json", "authorization": "Basic"}
```

```python
#report id for chapi room base
roomreportID = 70725
#report id for chapi events
eventreportID = 84711
```

```python
room_command = f"getreport/{roomreportID}.csv"
event_command  = f"getreport/{eventreportID}.csv"
```

```python
room_response = requests.get(starrez_url + room_command,headers=headers,auth = basic)
```

```python
event_response = requests.get(starrez_url+event_command,headers=headers,auth = basic)
```

```python
with open("conferences.csv",'w',encoding= 'utf-8') as events:
  events.write(event_response.text)
```

```python
with open("rooms.csv",'w',encoding ='utf-8') as rooms:
    rooms.write(room_response.text)
```
